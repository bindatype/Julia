```julia
function computePi(size)
    x=rand(size,2)
    count =0;
    for i ∈ 1:size 
        if +(x[i,1]^2,x[i,2]^2) <=1
            count+=1
       # println(count)
        end
    end
 return(4*count/size)
end
```
